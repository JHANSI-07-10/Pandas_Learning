# Pandas Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\HP\Downloads\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
df.plot(kind = 'line',title = "Ice cream Ratings",xlabel = 'Daily Ratings' , ylabel = "scores")
```

```python
df.plot(kind = 'bar',stacked = True,title = "Ice cream Ratings",xlabel = 'Daily Ratings' , ylabel = "scores")
```

```python
df['Flavor Rating'].plot(kind = 'bar')
```

```python
df.plot.scatter(x = 'Texture Rating',y = 'Overall Rating',s = 100 , c = "red")
```

```python
df.plot.hist(bins = 20)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (5,5))
```

```python
df.plot.pie(y = 'Texture Rating',figsize = (4,7))
```
